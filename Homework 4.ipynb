## Question 5

```python
marks = {'Andy':88, 'Amy':66, 'James': 90, 'Jules': 55, 'Arthur': 77}
marks
```

```python
names = marks.keys()
grades = marks.values()

plt.bar(names, grades)
plt.title('Grades')
plt.show()
```

```python
plt.pie(grades, labels = names)
plt.legend(title = 'Students',bbox_to_anchor = (1.0, 1.0), loc = 'upper left')
plt.title('Pie Chart of Grades')
plt.show()
```

## Question 6

```python
# PLOT 1 - MULTIPLE LINES
y1 = np.arange(7)
y2 = np.arange(3,10)
y3 = np.arange(6,13)

plt.subplot(2,3,1)
plt.plot(y1, color = 'r')
plt.plot(y2)
plt.plot(y3, color = 'g')
plt.title('Multiple Lines')

#PLOT 2 - BAR CHART
x = np.array([1,3,5,7,9])
y = ['A','B','C','D','E']

plt.subplot(2,3,2)
plt.bar(x,y)
plt.title('Bar Chart')

#PLOT 3 - PIE
x = np.array([10,40,70,30])

plt.subplot(2,3,3)
plt.pie(x)
plt.title('Pie Chart')

#PLOT 4 - GRID
x = np.arange(0,80)
y = np.arange(100,20,-1)

plt.subplot(2,3,4)
plt.plot(x,y)
plt.title('Grid')
plt.grid(color = 'r', lw = 0.25)

#PLOT 5 - SCATTER
x = np.array([56,4,78,35,66,10,9,28])
y = np.array([3,65,42,31,11,23,6,8])

plt.subplot(2,3,5)
plt.scatter(x,y)
plt.title('Scatter Plot')

#PLOT 6 - HISTOGRAM
x = np.random.normal(170,10,500)

plt.subplot(2,3,6)
plt.hist(x)
plt.title('Histogram')

plt.tight_layout(pad = 2.0)
plt.show()
```
